# single

### cores_close

```python
df_cores_close_mkl=pd.read_csv(r"/home/davide/università/hpc/project/hpc-project/exercise2/weak_scalability/single/cores_close_mkl.csv", decimal=',')
df_cores_close_oblas=pd.read_csv(r"/home/davide/università/hpc/project/hpc-project/exercise2/weak_scalability/single/cores_close_oblas.csv", decimal=',')

cores_close_mkl=df_cores_close_mkl.groupby('size').agg({'Gflops': ['mean', 'std']})
cores_close_mkl=cores_close_mkl.xs('Gflops', axis=1, drop_level=True)
cores_close_mkl=cores_close_mkl.reset_index('size')
cores_close_mkl.rename(columns={"mean":"Gflops"}, inplace=True)



cores_close_oblas=df_cores_close_oblas.groupby('size').agg({'Gflops': ['mean', 'std']})
cores_close_oblas=cores_close_oblas.xs('Gflops', axis=1, drop_level=True)
cores_close_oblas=cores_close_oblas.reset_index('size')
cores_close_oblas.rename(columns={"mean":"Gflops"}, inplace=True)




ax=cores_close_mkl.plot(kind='line', x='size', y='Gflops', yerr='std', linewidth=1, marker=('o'), color='r', label="mkl")
cores_close_oblas.plot(ax=ax, kind='line', x='size', y='Gflops', yerr='std', linewidth=1, marker=('o'), color='b', label="oblas")


#dimension and name of axis
plt.xlabel("Size", fontsize=15, alpha=0.5)
plt.ylabel("GFlops", fontsize=15,  alpha=0.5)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.tight_layout()

plt.legend()

plt.show()
```
