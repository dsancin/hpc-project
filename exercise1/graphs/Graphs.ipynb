```python
import numpy as np
import matplotlib.pyplot as plt
```

### openmp strong scalability

```python
x=[1, 2, 4, 8, 12, 16, 24, 32, 48, 64, 96, 128]
y_5000=[1, 1.92, 3.62, 6.26, 7.97, 9.45, 12.92, 14.54, 17.95, 19.34, 11.78, 14.20]
y_10000=[1, 1.93, 3.60, 6.52, 9.17, 11.18, 14.44, 17.20, 20.38, 22.44, 17.54, 17.57]
y_ideal=[1, 2, 4, 8, 12, 16, 24, 32, 48, 64, 96, 128]

plt.plot(x,y_5000,linestyle='-',linewidth=1,color='r',marker=('o'), label="5000")
plt.plot(x,y_10000,linestyle='-',linewidth=1,color='b',marker=('o'), label="10000")
plt.plot(x,y_ideal,linestyle='-',linewidth=1,color='g',marker=(' '), label="Ideal")

#dimension and name of axis
plt.xlabel("OpenMP threads", fontsize=15, alpha=0.5)
plt.ylabel("Speedup", fontsize=15,  alpha=0.5)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.tight_layout()

plt.legend()

plt.show()
```
